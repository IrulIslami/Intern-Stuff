```python
# def clean_kabupaten_name(name):
#     """
#     Cleans and standardizes kabupaten names by:
#     1. Converting to uppercase
#     2. Removing leading/trailing spaces
#     3. Removing common prefixes like 'KAB.', 'KABUPATEN', 'KOTA'
#     """
#     name = name.upper().strip()
#     # Remove prefixes. Use regex to handle spaces and dots after the prefix.
#     name = re.sub(r'^(KAB\.?|KABUPATEN)\s?', '', name)
#     return name

def clean_kabupaten_name(name):
    """
    Cleans and standardizes kabupaten names by:
    1. Converting to lowercase
    2. Removing all spaces
    3. Removing common prefixes like 'KAB.', 'KABUPATEN', 'KOTA'
    """
    name = str(name).lower().strip()
    # Remove prefixes. Use regex to handle spaces and dots after the prefix.
    name = re.sub(r'^(kab\.?|kabupaten|kota)\s?', '', name)
    # Remove all remaining spaces
    name = re.sub(r'\s+', '', name)
    return name

# Apply the cleaning function to both DataFrames
import re

df_internet['kabupaten_clean'] = df_internet['KABUPATEN'].apply(clean_kabupaten_name)
df_kode['kabupaten_clean'] = df_kode['kabupaten'].apply(clean_kabupaten_name)

print("\nDataFrame 1 with cleaned names:")
print(df_internet[['KABUPATEN', 'kabupaten_clean']])
print("\nDataFrame 2 with cleaned names:")
print(df_kode[['kabupaten', 'kabupaten_clean']])
```

```python
# Find the set difference
missing_kabupaten = set(df_kode['kabupaten_clean']) - set(df_internet['kabupaten_clean'])

print("Kabupaten in the reference list (df2) not found in your main data (df1):")
missing_kabupaten
```

```python
print(len(missing_kabupaten))
```

```python
# --- PERFORM THE LOOKUP/MERGE ---
# Create a mapping dictionary from the reference dataframe
kode_mapping = df_kode.set_index('kabupaten_clean')['kode'].to_dict()

# Map the kode values to the main dataframe
df_internet['Kode_Kabupaten'] = df_internet['kabupaten_clean'].map(kode_mapping)

# --- CHECK FOR MISSING MATCHES ---
missing_mask = df_internet['Kode_Kabupaten'].isna()
missing_count = missing_mask.sum()

print("\n" + "="*60)
print("MERGE RESULTS:")
print("="*60)
print(f"Successfully matched: {len(df_internet) - missing_count} kabupaten")
print(f"Failed to match: {missing_count} kabupaten")

if missing_count > 0:
    print("\nKabupaten that could not be matched (no kode found):")
    print(df_internet[missing_mask]['KABUPATEN'].tolist())

# --- FINAL DATAFRAME ---
# You can drop the temporary clean column if you want
df1_final = df_internet.drop('kabupaten_clean', axis=1)

print("\n" + "="*60)
print("FINAL DATAFRAME WITH KODES:")
print("="*60)
print(df1_final)
```

```python
column_order = ['Kode_Kabupaten', 'PROVINSI', 'KABUPATEN', '5G/4G/LTE_2021', 
                '3G/H/H+/EVDO_2021', '2,5G/E/GPRS_2021', '5G/4G/LTE_2024', 
                '3G/H/H+/EVDO_2024', '2,5G/E/GPRS_2024']

# Only include columns that actually exist in your dataframe
column_order = [col for col in column_order if col in df1_final.columns]
df1_final = df1_final[column_order]

print(df1_final)
```

```python
df1_final.to_csv('telecom_data_with_kode_kabupaten_irul.csv', index=False)
```
