```python
import os
import pandas as pd
import glob

def merge_csv_in_pelita_directory():
    # Path to the pelita directory
    pelita_dir = 'pelita'
    
    # Check if the directory exists
    if not os.path.exists(pelita_dir):
        print(f"Directory '{pelita_dir}' not found!")
        return
    
    # Get all CSV files in the pelita directory
    files = glob.glob(os.path.join(pelita_dir, 'pelita(*).csv'))
    
    # If no files found with that pattern, try alternative patterns
    if not files:
        files = glob.glob(os.path.join(pelita_dir, '*.csv'))
        print("Trying alternative pattern...")
    
    # Sort files numerically
    files.sort(key=lambda x: extract_number(x))
    
    print(f"Found {len(files)} files in '{pelita_dir}' directory:")
    for file in files:
        print(f"  - {os.path.basename(file)}")
    
    if files:
        # Read and merge all CSV files
        dfs = []
        for file in files:
            try:
                df = pd.read_csv(file)
                dfs.append(df)
                print(f"âœ“ Successfully read {os.path.basename(file)} ({len(df)} rows)")
            except Exception as e:
                print(f"âœ— Error reading {os.path.basename(file)}: {e}")
        
        if dfs:
            merged_df = pd.concat(dfs, ignore_index=True)
            output_file = 'merged_pelita.csv'
            merged_df.to_csv(output_file, index=False)
            print(f"\nâœ… Successfully merged {len(dfs)} files into {output_file}")
            print(f"ðŸ“Š Total rows: {len(merged_df)}")
        else:
            print("No files were successfully read!")
    else:
        print(f"No CSV files found in '{pelita_dir}' directory!")

def extract_number(filename):
    """Extract number from filename"""
    import re
    match = re.search(r'\((\d+)\)', filename)
    if match:
        return int(match.group(1))
    return 0  # default if no number found

# Run the function
merge_csv_in_pelita_directory()
```
